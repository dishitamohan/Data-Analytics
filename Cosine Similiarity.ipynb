# 1.Summary of MovieLens Metadata:

This dataset was generated on September 26, 2018.

1.Contains 100836 ratings and 3683 tag applications across 9742 movies

2.Users selected at random for inclusion. All selected users had rated at least 20 movies

3.Ratings are made on a 5-star scale, with half-star increments (0.5 stars - 5.0 stars).

4.Timestamps represent seconds since midnight Coordinated Universal Time (UTC) of January 1, 1970.

5.Movie titles are entered manually or imported from https://www.themoviedb.org/, and include the year of release in parentheses. Errors and inconsistencies may exist in these titles.

6.Genres are a pipe-separated list, and are selected from the following:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
import datetime

from sklearn.metrics import pairwise_distances as pair
from scipy.spatial.distance import cosine, correlation


df1 = pd.read_csv('links.csv')
df2 = pd.read_csv('movies.csv')
df3 = pd.read_csv('ratings.csv')
df4 = pd.read_csv('tags.csv')
```

```python
df=df1.merge(df2, left_on='movieId', right_on='movieId', how='left')
df=df.merge(df3, left_on='movieId', right_on='movieId', how='left')
df=df.merge(df4, left_on='movieId', right_on='movieId', how='left')
df
```

```python
#2. Read the “ratings.csv” file and create a pivot table with index=‘userId’, columns=‘movieId’, values = “rating".

dff = df3.pivot_table(index='userId', columns='movieId', values='rating',fill_value=0, aggfunc=np.mean)

dff
```

```python
#3. sklearn.metrics.pairwise_distances can be used to compute distance between all pairs of 
#users. pairwise_distances() takes a metric parameter for what distance measure to use. 
#Use cosine similarity for finding similarity among users. Use the following packages

pair=pairwise_distances(dff, Y=None, metric='cosine')
pair
```

```python
#6. Find the 5 most similar user for user with user Id 25.

print((-pair[24]).argsort()[1:6]+1)
```

```python
#7. Use the “movies” dataset to find out the names of movies, user 1 and user 338 have watched in common and how they have rated each one of them.

common_movies = set(df3.loc[df3.userId==1, 'movieId']).intersection(set(df3.loc[df3.userId==338, 'movieId']))
common_movies
```

```python
#8. Use the movies dataset to find out the common movie names between user 2 and user 338 with least rating of 4.0


# common_movies2 = set(df3.loc[df3.userId==2, 'movieId']).intersection(set(df3.loc[df3.userId==338, 'movieId']))
common_movies2 = set(df3.loc[((df3.userId==2) & (df3.rating>=4.0)), 'movieId']).intersection(set(df3.loc[((df3.userId==338) & (df3.rating>=4.0)), 'movieId']))

common_movies2


for movieid in sorted(common_movies2):
    common = df2[df2['movieId'] == movieid]
    print(common)
```

```python
#9. Create a pivot table for representing the similarity among movies using correlation.

dff = pd.pivot_table(df3, index='movieId', columns='userId', values='rating', fill_value=0)

pair2=pairwise_distances(dff, Y=None, metric='correlation')

pair2
```
